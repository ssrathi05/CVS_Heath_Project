```python
import pandas as pd
import os
```

```python
# load the full Chronic Disease Indicators dataset
cdi_df_full = pd.read_csv(
    r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\raw\U.S._Chronic_Disease_Indicators_20251203.csv",
    dtype={"FIPS": str},
    encoding="latin1",
    low_memory=False
)


```

```python
cdi_df_full.shape
```

```python
cdi_df_full["Topic"].unique()[:20]
```

```python
diabetes_df = cdi_df_full[cdi_df_full["Topic"] == "Diabetes"] 
diabetes_df.shape
```

```python


diabetes_df["DataValueType"].value_counts()


```

```python
diabetes_df = diabetes_df[diabetes_df["DataValueType"] == "Crude Prevalence"]
diabetes_df.shape
```

```python
diabetes_df.columns.tolist()

```

```python
# Create a proper FIPS column from LocationID
diabetes_df["FIPS"] = diabetes_df["LocationID"].astype(str).str.zfill(5)

# Now filter to keep only valid county FIPS rows (5-digit codes)
diabetes_df = diabetes_df[diabetes_df["FIPS"].str.len() == 5]
diabetes_df.head()
```

```python
# Find the most recent year in the filtered dataset
latest_year = diabetes_df["YearStart"].max()
latest_year
```

```python
#using latest year - aka 2022
# keep only the latest year rows
diabetes_df = diabetes_df[diabetes_df["YearStart"] == latest_year]
diabetes_df.shape
```

```python
diabetes_clean = diabetes_df[[
    "FIPS",
    "LocationAbbr",
    "LocationDesc",
    "DataValue"
]].rename(columns={
    "LocationAbbr": "ST",
    "LocationDesc": "County",
    "DataValue": "DiabetesPct"
})

diabetes_clean.head()
```

```python
output_path = r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\processed\diabetes_clean.csv"
diabetes_clean.to_csv(output_path, index=False)

output_path
```
