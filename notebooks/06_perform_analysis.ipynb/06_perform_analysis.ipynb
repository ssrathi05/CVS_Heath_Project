 # CVS Health Community Access Analysis  
This notebook analyzes county-level health needs, social vulnerability, preventive care rates, and MinuteClinic availability to identify underserved communities.  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\processed\CVS_FINAL_DATASET.csv")

df.head()

```

```python
# remove commas or percent signs if needed
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

# ensure COUNTY is str FIPS
df["COUNTY"] = df["COUNTY"].astype(str).str.zfill(5)
```

## High-Level Metrics


```python
print("Total counties:", len(df))
print("Counties with zero clinics:", sum(df['clinic_count'] == 0))
print("Average SVI:", df['svi_overall'].mean())
print("States with most high-need counties:")
df[df['clinic_count']==0].groupby('state_full').size().sort_values(ascending=False).head(10)
```

```python
state_summary = df.groupby("state_full").agg({
    "clinic_count": "sum",
    "need_score": "mean",
    "svi_overall": "mean"
})

state_summary.sort_values("need_score", ascending=False).head(15)
```
