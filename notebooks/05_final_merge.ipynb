```python
import pandas as pd
```

```python
# load cleaned datasets
clinic = pd.read_csv(r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\processed\cvs_minute_clinics_cleaned.csv")
svi_places = pd.read_csv(r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\processed\svi_places_wide_cleaned.csv")
```

```python
# ensure FIPS type matches
clinic['COUNTY'] = clinic['COUNTY'].astype(str).str.zfill(5)
svi_places['COUNTY'] = svi_places['fips'].astype(str).str.zfill(5)
```

```python
final_df = clinic.merge(
    svi_places,
    on="COUNTY",
    how="left"
)
```

```python
final_df.head()
```

```python
clinic_counts = clinic.groupby("COUNTY").size().reset_index(name="clinic_count")
clinic_counts.head()
```

```python
county_level_df = svi_places.merge(
    clinic_counts,
    on="COUNTY",
    how="left"
)

# fill counties with 0 clinics
county_level_df["clinic_count"] = county_level_df["clinic_count"].fillna(0)
```
