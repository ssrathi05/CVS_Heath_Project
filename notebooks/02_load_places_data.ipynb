```python
places_df = places_df_og[places_df_og["Data_Value_Type"] == "Crude prevalence"]
```

```python
places_df.head(20)
```

```python
#keep only crude prevalence rows
#next remove unneeded columns
places_df = places_df[[
    "StateAbbr",
    "LocationName",
    "Category",
    "Short_Question_Text",
    "Data_Value",
    "Geolocation"
]]
```

```python
places_df.head()  
```

```python
places_df["StateAbbr"] = places_df["StateAbbr"].str.upper().str.strip()
places_df["LocationName"] = places_df["LocationName"].str.upper().str.strip()
```

```python
places_df.head()
```

```python
fips_url = "https://raw.githubusercontent.com/kjhealy/fips-codes/master/state_and_county_fips_master.csv"
fips_df = pd.read_csv(fips_url)
```

```python
fips_df.head()
```

```python
fips_df = fips_df.dropna(subset=["state"])
```

```python
fips_df["county_clean"] = (
    fips_df["name"]
    .str.replace(" County", "", regex=False)
    .str.upper()
    .str.strip()
)
```

```python
fips_df
```

```python
fips_df["FIPS"] = fips_df["fips"].astype(int).astype(str).str.zfill(5)
```

```python
fips_df.head()
```

```python
fips_df["state"] = fips_df["state"].str.upper().str.strip()
```

```python
places_df = places_df.merge(
    fips_df[["FIPS", "state", "county_clean"]],
    left_on=["StateAbbr", "LocationName"],
    right_on=["state", "county_clean"],
    how="left"
)
```

```python
places_df.head()
```

```python
places_df = places_df.drop(columns=["state", "county_clean"])
```

```python
places_df.head()
```

```python
places_df.to_csv("places_cleaned.csv", index=False)
```
