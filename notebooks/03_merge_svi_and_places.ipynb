```python
import pandas as pd
```

```python
places = pd.read_csv(r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\processed\places_cleaned.csv", dtype={"FIPS": str})
svi = pd.read_csv(r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\processed\svi_cleaned.csv", dtype={"FIPS": str})
```

```python
merged_svi_places_df = places.merge(
    svi,
    on="FIPS",
    how="left"   # keep all PLACES rows
)
```

```python
merged_svi_places_df
```

```python
merged_svi_places_df.shape
```

```python
pivot_df = merged_svi_places_df[[
    "FIPS",
    "StateAbbr",
    "STATE",
    "COUNTY",
    "Short_Question_Text",
    "Data_Value",
 
    
    "RPL_THEMES",
    "RPL_THEME1",
    "RPL_THEME2",
    "RPL_THEME3",
    "RPL_THEME4"
]].copy()
```

```python
pivot_df.head()
```

```python
wide = pivot_df.pivot_table(
    index=[
        "FIPS",
        "StateAbbr",
        "STATE",
        "COUNTY",
        "RPL_THEMES",
        "RPL_THEME1",
        "RPL_THEME2",
        "RPL_THEME3",
        "RPL_THEME4"
    ],
    columns="Short_Question_Text",
    values="Data_Value"
).reset_index()
```

```python
wide
```

```python
# Clean all column names
wide.columns = (
    wide.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.replace("__", "_")
)
```

```python
wide = wide.rename(columns={
    "fips": "fips",
    "stateabbr": "state",
    "state": "state_full",
    "county": "county_full",
    "rpl_themes": "svi_overall",
    "rpl_theme1": "svi_socioeconomic",
    "rpl_theme2": "svi_household",
    "rpl_theme3": "svi_minority",
    "rpl_theme4": "svi_housing"
})
```

```python
wide
```

```python
location_cols = [
    "fips", "state", "state_full", "county_full"
]

svi_cols = [
    "svi_overall", "svi_socioeconomic", "svi_household", 
    "svi_minority", "svi_housing"
]

other_cols = [c for c in wide.columns if c not in location_cols + svi_cols]

wide = wide[location_cols + svi_cols + other_cols]
```
