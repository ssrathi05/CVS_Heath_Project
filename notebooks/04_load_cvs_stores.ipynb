```python
clinic = cvs_stores[cvs_stores["indicatorMinuteClinicService"] == "Y"]
```

```python
clinic.columns.tolist()
```

```python
clinic = clinic[[
    "StoreNumber",
    "minuteClinicID",
    "addressLine",
    "addressState",
    "addressZipCode",
    "geographicLatitudePoint",
    "geographicLongitudePoint"
]].copy()
```

```python
clinic
```

```python
clinic = clinic.rename(columns={
    "StoreNumber": "store_id",
    "minuteClinicID": "clinic_id",
    "addressLine": "address",
    "addressState": "state",
    "addressZipCode": "zip",
    "geographicLatitudePoint": "latitude",
    "geographicLongitudePoint": "longitude"
})

# Convert coordinates to numeric
clinic["latitude"] = pd.to_numeric(clinic["latitude"], errors="coerce")
clinic["longitude"] = pd.to_numeric(clinic["longitude"], errors="coerce")

clinic.head()
```

```python
zip_fips = pd.read_excel(r"C:\Users\14122\OneDrive\Desktop\cvs_heath_project\data\raw\ZIP_COUNTY_062025.xlsx")
```

```python
zip_fips['ZIP'] = zip_fips['ZIP'].astype(str).str.zfill(5)
```

```python
#merge data
clinic['zip'] = clinic['zip'].astype(str).str.zfill(5)
# merge
merged_clinic = clinic.merge(
    zip_fips[['ZIP', 'COUNTY']],
    left_on='zip',
    right_on='ZIP',
    how='left'
)
```
